```python
import os
import glob
import shutil
import random

# YOLO etiket ve görüntü klasörleri
images_dir = r"C:\Users\524ha\Desktop\datasets\all_annotated"  # Tüm resimlerin olduğu klasör
labels_dir = r"C:\Users\524ha\Desktop\datasets\all_annotated"  # Tüm etiketlerin olduğu klasör
train_img_dir = r"C:\Users\524ha\Desktop\datasets\train\images"
train_lbl_dir = r"C:\Users\524ha\Desktop\datasets\train\labels"
val_img_dir = r"C:\Users\524ha\Desktop\datasets\val\images"
val_lbl_dir = r"C:\Users\524ha\Desktop\datasets\val\labels"

# Train/Val oranı
train_ratio = 0.8

# Klasörleri oluştur
os.makedirs(train_img_dir, exist_ok=True)
os.makedirs(train_lbl_dir, exist_ok=True)
os.makedirs(val_img_dir, exist_ok=True)
os.makedirs(val_lbl_dir, exist_ok=True)

# Tüm görüntü dosyalarını listele (.jpg ve .png)
image_files = glob.glob(os.path.join(images_dir, "*.jpg")) + glob.glob(os.path.join(images_dir, "*.png"))

# Dosyaları rastgele sırala
random.shuffle(image_files)

# Train/Val bölme noktası
split_idx = int(len(image_files) * train_ratio)

# Dosyaları ilgili klasörlere taşı
for i, img_file in enumerate(image_files):
    # Aynı isimdeki .txt etiket dosyasını bul
    base_name = os.path.basename(img_file).rsplit(".", 1)[0]  # Dosya adını uzantısız al
    label_file = os.path.join(labels_dir, base_name + ".txt")

    # Train klasörüne taşı
    if i < split_idx:
        shutil.copy(img_file, train_img_dir)
        if os.path.exists(label_file):
            shutil.copy(label_file, train_lbl_dir)
    # Val klasörüne taşı
    else:
        shutil.copy(img_file, val_img_dir)
        if os.path.exists(label_file):
            shutil.copy(label_file, val_lbl_dir)

print(f"Toplam {len(image_files)} görüntü bulundu.")
print(f"Train set: {split_idx} görüntü")
print(f"Val set: {len(image_files) - split_idx} görüntü")
```
