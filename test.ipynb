```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
model = YOLO("mini_tube_detector_model.pt")
```

```python
frame = cv2.imread("image.jpg")
frame_ = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
result = model(frame_)
frame_ = result[0].plot()
plt.imshow(frame_)
```
